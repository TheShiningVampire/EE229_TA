### Generating a Signal Waveform

#### Sine Wave

```python
x = np.linspace(0,3*np.pi, 100)
y = np.sin(x)

plt.plot(x,y, label='sine wave')
plt.title('sine wave')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.grid(True)
plt.legend()
```

#### Real Exponential Signal

```python
x = np.linspace(0,3*np.pi, 100)
y = np.exp(x)

plt.plot(x,y, label='exp signal')
plt.title('Exp Signal')
plt.xlabel('x')
plt.ylabel('exp(x)')
plt.grid(True)
plt.legend()
```

#### Complex Exponential Wave
Generating a complex exponential waveform is a bit more involved than a sine wave. We want to plot the following function:
$$ y(t) = A e^{\alpha t} $$
for  A = 1
and    
    1) $\alpha$ = j*50  
    2) $\alpha$ = -1 + j*50



#### Expected results for 1

Real Part of the signal  
<img src="1a.png" width="500" height="300">   <br>
Imaginary Part of the signal  
<img src="1b.png" width="500" height="300">  

Magnitude of the signal  
<img src="1c.png" width="500" height="300">  

```python
A = 1
omega = 50
x = np.linspace(0,0.5, 10000)
# Complex exponential function
y = A*np.exp(1j*omega*x)

# Plot the magnitude of the complex exponential function
plt.figure(figsize=(10,6))
plt.plot(x,np.real(y))
plt.title('Real part')
plt.grid(True)
plt.show()
plt.figure(figsize=(10,6))
plt.plot(x,np.imag(y))
plt.title('Imaginary part')
plt.grid(True)
plt.show()
plt.figure(figsize=(10,6))
plt.plot(x,np.abs(y))
plt.title('Magnitude')
plt.grid(True)
plt.show()
```
